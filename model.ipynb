```python
X = data.drop(columns=['is_fraud'])
y = data['is_fraud']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, mean_absolute_error, roc_auc_score

# Load the dataset
# Assuming data is already loaded into a variable called 'data'

# Separate features and target variable
X = data.drop(columns=['is_fraud'])
y = data['is_fraud']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, columns=['merchant', 'category', 'city', 'state', 'job'])

# Drop columns 'dob' and 'trans_date' if they are not needed for training
X_encoded.drop(columns=['dob', 'trans_date'], inplace=True)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Apply SMOTE to balance the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Train the Random Forest Classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))


accuracy = accuracy_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Mean Absolute Error:", mae)
print("ROC-AUC Score:", roc_auc)
```
